```python
#importint all the neccessary packages to be used 
import pandas as pd 
import numpy as np
import seaborn as sns
```

```python
#reading the data set using pandas
df_FBS = pd.read_csv("FoodBalanceSheet_E_NOFLAG.csv", encoding='latin-1')
```

```python
#getting the first 5 rows of the dataset
df_FBS.head()
```

```python
#getting an overview of the data to get the unique number, mean and standard deviation
df_FBS.describe(include = 'all')
```

```python
#checking for missing values
df_FBS.isnull().sum()
```

```python
#calling formula to calculate the missing value percentage for each attribute
missing_values_percentage = df_FBS.isnull().mean()*100
```

```python
# to pring out missing formula
print(missing_values_percentage)
```

```python
#calling for the total count of each column that are not null
df_FBS.count()
```

```python
#summing up all observation in 2014
df_FBS.Y2014.sum()
```

```python
#summing up all observation in 2018
df_FBS.Y2018.sum()
```

```python
df_FBS.groupby('Element').sum()
```

```python
#total protein supply quantity in Madagascar in 2015
df_FBS[(df_FBS['Element'] == 'Protein supply quantity (g/capita/day)') & (df_FBS['Area'] == 'Madagascar')].groupby(['Area', 'Element'])['Y2015'].sum().sum()
```

```python
#groupby operation on element to calculate the total number of the sum of processing in 2017
df_FBS[(df_FBS['Element'] == 'Processing')].groupby(['Element'])['Y2017'].sum().sum()
```

```python
#a groupby operation to determine the area with the highest sum in 2017
df_FBS[['Y2017', 'Area']].groupby(['Area']).sum().sort_values(by='Y2017', ascending=False).reset_index().iloc[0]['Area']
```

```python
#groupby operation on area to determine the areas that had the 7th lowest sum in 2017
df_FBS[['Y2017', 'Area']].groupby(['Area']).sum().sort_values(by='Y2017').reset_index().iloc[6]['Area']
```

```python
#to calculate the year with the least correlation with 'Element Code' column
df_FBS[['Element Code', 'Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']].corr()['Element Code'].sort_values().index[0]
```

```python
#groupby operation on 'Element' to get the year with the highest sum of stock variation
df_FBS[df_FBS['Element'] == 'stock variation'].groupby(['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']).sum().sort_values(by='Element', ascending=True).reset_index().iloc[0][0]
```
